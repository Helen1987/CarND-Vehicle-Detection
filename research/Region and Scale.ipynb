```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

import sys
sys.path.append('../')
from src.WindowSlider import WindowSlider
%matplotlib inline

project_dir = os.path.dirname(os.getcwd())
test_images_folder_path = os.path.join(project_dir, 'test_images')

path = os.path.join(test_images_folder_path, 'test1.jpg')
img = cv2.imread(path)
slider = WindowSlider(None, None, None)

def draw_region(img, scale, x_start_stop, y_start_stop, xy_overlap):
    draw_image = np.copy(img)
    win_draw = np.int(64*scale)
    scaled_image = slider.get_scaled_image_region(img, scale, x_start_stop, y_start_stop)
    windows = slider.slide_window(scaled_image.shape, xy_overlap=xy_overlap)

    for (x_pos, y_pos), (x_start, y_start) in windows:
        x_top_left = np.int(x_start*scale)+x_start_stop[0]
        y_top_left = np.int(y_start*scale)+y_start_stop[0]

        cv2.rectangle(
            draw_image, (x_top_left, y_top_left),
            (x_top_left+win_draw, y_top_left+win_draw),
            (0,0,255), 2) 
    
    rgb_image = cv2.cvtColor(draw_image, cv2.COLOR_BGR2RGB)    
    plt.figure()
    plt.imshow(rgb_image)  
```

```python
scale = 1
x_start_stop = (300, img.shape[1]-300)
y_start_stop = (380, 450)
xy_overlap=(0.75, 0.75)


draw_region(img, scale, x_start_stop, y_start_stop, xy_overlap) 
```
